```python
import numpy as np 
import neural_helper as helper
import matplotlib.pyplot as plt
```

```python
# Create a grid
x = np.linspace(0,10, 100)
y = np.linspace(0,10, 100)
X, Y = np.meshgrid(x, y)
# Apply periodic potential
V = helper.potential(X, 10,15,1)

# Plotting
plt.figure(figsize=(8, 4))
plt.pcolormesh(X, Y, V, shading='auto', cmap='viridis')
plt.colorbar(label='Potential V(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('1D Periodic Potential in 2D Grid')
plt.show()
```
